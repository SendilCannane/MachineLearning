```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#day=['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
day=[1,2,3,4,5,6,7]
Max_temp=[85,75,45,35,25,65,55]
Min_temp=[20,30,40,60,50,70,45]
Avg_temp=[15,24,36,56,76,83,32]

plt.plot(day,Max_temp,'ro--',label='Max')
plt.plot(day,Min_temp,'go--',label='Min')
plt.plot(day,Avg_temp,'bo--',label="Avg")
plt.legend() 
plt.grid()
```
